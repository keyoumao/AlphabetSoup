```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

# Import our input dataset
ASC_df = pd.read_csv('charity_data.csv')
ASC_df.head()
```

```python
# Generate our categorical variable list
ASC_cat = ASC_df.dtypes[ASC_df.dtypes == "object"].index.tolist()
```

```python
# Check the number of unique values in each column
ASC_df[ASC_cat].nunique()
```

```python
ASC_df=ASC_df.drop(columns=['EIN','CLASSIFICATION','APPLICATION_TYPE','STATUS'])
# Generate our categorical variable list
ASC_cat = ASC_df.dtypes[ASC_df.dtypes == "object"].index.tolist()
```

```python
name_counts = ASC_df.NAME.value_counts()
name_counts
```

```python
# Visualize the value counts
name_counts.plot.density()
```

```python
# Determine which values to replace
replace_names = list(name_counts[name_counts < 100].index)
delete_names = list(name_counts[name_counts < 20].index)

for com in delete_names:
    ASC_df = ASC_df[ASC_df.NAME != com]

# Replace in DataFrame
for name in replace_names:
    ASC_df.NAME = ASC_df.NAME.replace(name,"Other")


# Check to make sure binning was successful
ASC_df.NAME.value_counts()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(ASC_df[ASC_cat]))

# Add the encoded variable names to the DataFrame
encode_df.columns = enc.get_feature_names(ASC_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
ASC_df = ASC_df.merge(encode_df,left_index=True, right_index=True)
ASC_df = ASC_df.drop(ASC_cat,1)
ASC_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = ASC_df["IS_SUCCESSFUL"].values
X = ASC_df.drop(["IS_SUCCESSFUL"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net
# We try two layers with 20 and 10 nodes first, and then 
# we add one more layer
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  20
hidden_nodes_layer2 = 10
hidden_nodes_layer3 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
